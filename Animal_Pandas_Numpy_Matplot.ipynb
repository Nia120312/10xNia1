```python
import pandas as pd


#replace 'your_dataset.csv' with the actual file path
animal = 'animal.csv'

#read csv file into dataframe
df = pd.read_csv(animal)

#display the first few rows of the dataframe 
print(df.head())

#display all rows of dataframe = print(df)
#to print in table form, type: df

```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_types = df.dtypes
print("Data Types of Each Coloumn: ")
print (data_types)

```

```python
df
```

```python
animal_names = df['Animal']
print("Names of all the animals")
print(animal_names)
```

```python
habitat = input("Enter the habitat to search for: ")
animal_hab = df[df['Habitat'] ==habitat]
print(f"Animals with habitat {habitat}: ")
print(animal_hab[['Animal', 'Type', 'Diet']])
```

```python
age_limit = int(input("Enter the Age Limit: "))
older_animals = df[df['Lifespan'] > age_limit]
print(f"Animals who live more than {age_limit} years: ")
print(older_animals[['Animal', 'Type', 'Lifespan']])
    
```

```python
sorted_by_age = df.sort_values(by='Lifespan', ascending=True)
print("All animals sorted by age: ")
print(sorted_by_age[['Animal', 'Type', 'Habitat', 'Lifespan']])
```

```python
unique_hab = df['Habitat'].unique()
print("Unique values in the 'Habitat' coloumn: ")
print(unique_hab)
```

```python
hab_count = df['Habitat'].value_counts()
print("Count of occurences for each habitat: ")
print(hab_count)
```

```python
plt.figure(figsize=(10, 10))
hab_counts = df['Habitat'].value_counts()
plt.pie(hab_counts, labels=hab_counts.index, autopct='%1.1f%%', startangle=90, colors = ['red','orange','yellow', 'green','lightblue','blue','purple'])
plt.title("Distribiution of Habitat of Animals")
plt.show
```

```python
plt.figure(figsize=(10, 10))
diet_counts = df['Diet'].value_counts()
plt.pie(diet_counts, labels=diet_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lavender'])
plt.title("Distribiution of Diet of Animals")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
hab_counts = df['Habitat'].value_counts()
hab_counts.plot(kind='bar', color='lavender')
plt.xlabel("Habitat")
plt.ylabel("Number of Animals")
plt.title("Number of Animals vs Habitat")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
